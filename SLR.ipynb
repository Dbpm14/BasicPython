###Q1.  What is Simple Linear Regression (SLR)? Explain its purpose.


$Soln$: Simple Linear Regression (SLR) is a statistical method used to model the linear relationship between two quantitative variables: one independent variable ($X$) and one dependent variable ($Y$). Its primary purpose is to predict the value of the dependent variable based on the value of the independent variable and to evaluate the strength of that relationship.

The primary purpose of SLR are as
follows:


*   To forecast the value of the dependent variable based on a specific value of the independent variable. For example, predicting a person's weight based on their height.
*   To determine the strength and direction of the association between variables. Does $Y$ increase as $X$ increases (positive correlation), or does it decrease (negative correlation)
* To understand how a variable changes over time or across different levels of another factor.



###Q2. What are the key assumptions of Simple Linear Regression?

$Soln$: The key assumptions of SLR are as follows:
* The relationship between the independent variable ($X$) and the dependent variable ($Y$) must be linear. This means that a change in $X$ should result in a proportional change in $Y$, which can be represented by a straight line.
* The observations (data points) must be independent of each other. This means the value of one observation should not influence or predict the value of another. This is particularly important in time-series data, where "autocorrelation" can occur.
* For any given value of $X$, the errors (residuals) should be normally distributed. While the $X$ and $Y$ variables themselves don't necessarily have to be normal, the "leftover" error after fitting the line should follow a bell curve.
* The variance of the error terms must be constant across all levels of the independent variable. In simpler terms, the "spread" of the data points around the regression line should be roughly the same from the beginning of the line to the end.

###Q3. Write the mathematical equation for a simple linear regression model and explain each term.

$Soln$: The mathematical equation for SLR is:
$$Y = \beta_0 + \beta_1X + \epsilon$$
* $Y$ (Dependent Variable): This is the target or the outcome you are trying to predict or explain (e.g., Exam Score).
* $X$ (Independent Variable): This is the predictor or the feature you are using to make the prediction (e.g., Study Hours).
* $\beta_0$ (Intercept): This is a constant that represents the value of $Y$ when $X$ is zero. Geometrically, it is the point where the regression line crosses the Y-axis.
* $\beta_1$ (Slope Coefficient): This represents the sensitivity of $Y$ to changes in $X$. It tells you how much $Y$ is expected to increase (if positive) or decrease (if negative) for every one-unit increase in $X$.
* $\epsilon$ (Error Term/Residual): Also known as the "noise," this term accounts for the variation in $Y$ that cannot be explained by $X$. Since real-world data points rarely fall perfectly on a straight line, $\epsilon$ represents the distance between the actual observed data point and the predicted line.

###Q5.  What is the method of least squares in linear regression?

$Soln $: The Method of Least Squares is the standard mathematical technique used to determine the "line of best fit" for a regression model. Since data points in the real world rarely form a perfectly straight line, this method finds the specific line that minimizes the overall discrepancy between the actual data and the model's predictions.

The goal is to find the values for the intercept ($\beta_0$) and the slope ($\beta_1$) that make the line as close to all the data points as possible.
* For every data point, the model calculates the "residual" (the vertical distance between the actual point and the line).
* Each residual is squared. This serves two purposes: it ensures all values are positive (so they don't cancel each other out) and it gives more "weight" to outliers (larger errors).
* The method adjusts the line until the Sum of Squared Residuals (SSR) is at its absolute minimum.

###Q6. What is Logistic Regression? How does it differ from Linear Regression?


$Soln$: Logistic Regression models the relationship between independent variables and a categorical dependent variable. In its most common form (Binary Logistic Regression), the outcome has only two possible classes, such as:

Yes vs. No

Spam vs. Not Spam

Pass vs. Fail

Instead of fitting a straight line, Logistic Regression uses the Sigmoid Function (also called the Logistic Function) to map any real-valued number into a value between 0 and 1.

If we try to fit a straight line to categorical data (0s and 1s), the line will often predict values below 0 or above 1, which makes no sense for probabilities. The Sigmoid curve "squashes" the output to ensure it always stays within the valid $0$ to $1$ range.

###Q7.  Name and briefly describe three common evaluation metrics for regression models.
$Soln$:
1. **Mean Absolute Error (MAE)**:
  
    MAE is the average of the absolute differences between the actual observed values and the values predicted by the model. It tells you, on average, how far off your predictions are.
  
    It is easy to interpret because it is in the same units as your dependent variable.

2. **Mean Squared Error (MSE)**:

    MSE is the average of the squared differences between the actual and predicted values. By squaring the errors, this metric heavily penalizes larger errors (outliers) more than smaller ones.

    It is mathematically useful for optimization (like in the Method of Least Squares), though it is harder to interpret because the units are squared
    
3. **R-Squared ($R^2$)**:

    Also known as the Coefficient of Determination, $R^2$ measures the proportion of the variance in the dependent variable that is predictable from the independent variable(s). It usually ranges from 0 to 1.

    It provides a "goodness-of-fit" measure. An $R^2$ of 0.85 means that 85% of the variation in your data is explained by the model, while 15% is due to unexplained noise.

###Q8. What is the purpose of the R-squared metric in regression analysis?

$Soln$: The R-squared ($R^2$) metric serves the following purpose:

1. Measuring "Goodness-of-Fit"

    The primary purpose of $R^2$ is to quantify how well the regression line approximates the real data points. It tells us what percentage of the "movement" in your target variable ($Y$) can be attributed to our predictor ($X$).
    * $R^2 = 1$: The model explains 100% of the variance (perfect fit).
    * $R^2 = 0$: The model explains none of the variance; it’s no better than simply predicting the average value of $Y$.
2. Comparing Models
  
    $R^2$ provides a standardized scale (usually 0 to 1) that allows us to compare different models. For instance, if we are predicting "House Price" and Model A has an $R^2$ of 0.70 while Model B has 0.85, Model B is generally considered to have a better explanatory power.
3. Identifying the Gap (Noise vs. Signal)

    It helps us understand how much of our data is "unexplained." If our $R^2$ is 0.80, it implies that:
    * 80% of the variance is explained by your model (Signal).
    * 20% of the variance is due to random error or factors not included in your model (Noise).

###Q9. Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.


```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt


X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 5, 4, 5])

model = LinearRegression()
model.fit(X, y)

slope = model.coef_[0]
intercept = model.intercept_

print(f"Slope (beta_1): {slope}")
print(f"Intercept (beta_0): {intercept}")

y_pred = model.predict(X)

plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', linewidth=2, label='Regression Line')

plt.title('Simple Linear Regression Fit')
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent Variable (y)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

###Q10. How do you interpret the coefficients in a simple linear regression model?

$Soln$:
1. The Intercept ($\beta_0$)

   The intercept is the expected value of $Y$ when $X$ is zero.
    * Geometrically: It is where the regression line crosses the vertical y-axis.
    * Practical Meaning: In some cases, this has a real-world meaning (e.g., the base price of a house before adding square footage). In other cases, it is purely a mathematical constant used to position the line (e.g., a person's weight when their height is zero which is physically impossible).
2. The Slope ($\beta_1$)

    The slope is the most critical part of the model. It represents the average change in $Y$ for every one-unit increase in $X$.
    * Positive Slope: If $\beta_1$ is positive, $Y$ increases as $X$ increases (Direct relationship).
    * Negative Slope: If $\beta_1$ is negative, $Y$ decreases as $X$ increases (Inverse relationship).
    * Zero Slope: If $\beta_1$ is zero, there is no linear relationship; changes in $X$ do not affect $Y$.
