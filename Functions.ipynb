###6. Explain the concept of generators in Python and how they are defined.
-Generator functions are a concise and memory-efficient way to create iterators. They allow you to iterate over large sequences of data without loading the entire sequence into memory at once.



They are defined like regular functions but use the yield keyword instead of return. When a generator encounters yield, it returns a value and pauses its execution state, resuming from that exact spot the next time it is called..

###7. What are the advantages of using generators over regular functions?
-Generators do not store all values in memory at once; they generate and consume each value one at a time, which is ideal for large datasets.




 Because they produce values one by one, they can be used to represent infinite sequences (like a Fibonacci sequence) that would otherwise crash a program if stored in a list

###9. Explain the purpose and usage of the `map()` function in Python
-The map() function is a built-in tool that applies a specified function to every item in an iterable like a list or tuple.

It takes two main parameters: the function to apply and the iterable to process. It returns a map object an iterator, which can then be converted into a usable form like a list or tuple to see the results. For example, it can be used to square every number in a list or capitalize a list of strings

###10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
-The map() function is a built-in function that applies a given function to all the items in an iterable
[Ex: String, List] and returns an iterable map object (an iterator) of the results.

The reduce() function iterates over a collection like an array or list and applies a callback function to each element, accumulating the results into a single, final value, effectively reducing the collection to one output, such as a sum or product.

The filter() function is a built-in function that extracts elements from an iterable like a list, tuple, or set for which a specified function returns a True value. It applies a condition to each element and keeps only those that satisfy it.


```python
#Example of map()
numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x**2, numbers))
print(squared_numbers)
```

```python
#Example of reduce()
from functools import reduce
nums= [1,2,3,4]
sq_nums=reduce(lambda x: x**2, nums)
print(sq_nums)
```

```python
#Example of filter()
l=[1,2,3,4,5,6,7,8,9]
num=list(filter(lambda x: x%2==0,l))
print(num)
```

###1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

```python
a = list(map(int, input("Enter numbers by giving space").split()))

def sum_of_even_nums(numbers_list):
  current_even_sum = 0
  for num in numbers_list:
    if num % 2 == 0:
      current_even_sum += num
  return current_even_sum

total_even_sum = sum_of_even_nums(a)
print(total_even_sum)
```

###5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
def fibonacci_generator(n):
  a,b=0,1
  for i in range(n):
    yield a
    a,b=b,a+b

for num in fibonacci_generator(10):
  print(num)
```

###6. Write a generator function in Python that yields the powers of 2 up to a given exponent

```python
def power_of_two(exponent):
  for i in range(exponent+1):
    yield 2**i

for power in power_of_two(5):
  print(power)
```

###8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

```python
def sort_tuple_list(tuple_list):
  tuple_list.sort(key=lambda x: x[1])
  return tuple_list

tuple_list = [(1, 5), (3, 2), (2, 8), (4, 1)]
sorted_list = sort_tuple_list(tuple_list)
print(sorted_list)
```

###10.Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
string=input("Enter a string: ")
def remove_vowels(string):
  vowel=["a","e","i","o","u","A","E","I","O","U"]
  return "".join(filter(lambda x: x not in vowel,string))

remove_vowels(string)
```
