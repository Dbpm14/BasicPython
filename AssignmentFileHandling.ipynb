###6.  Write a Python program that logs an error message to a log file when a division by zero exception occurs.

```python
import logging

logging.basicConfig(filename='error_log.txt', level=logging.ERROR, format='%(asctime)s - %(levelname)s - %(message)s')

def divide_numbers(a, b):
    try:
        result = a / b
        return result
    except ZeroDivisionError as e:
        logging.error(f"Division by zero attempted. Inputs: a={a}, b={b}. Error: {e}")
        print("An error occurred. Please check the log file.")

divide_numbers(10, 0)
```

###8. Write a program to handle a file opening error using exception handling.

```python
def open_and_read_file(file_name):
    try:

        with open(file_name, 'r') as file:
            content = file.read()
            print("--- File Content ---")
            print(content)

    except FileNotFoundError:
        print(f"Error: The file '{file_name}' does not exist.")


    except Exception as e:

        print(f"An unexpected error occurred: {e}")

filename = input("Enter the name of the file to open: ")
open_and_read_file(filename)
```

###14.   Write a program that uses the logging module to log both informational and error messages

```python
import logging
import os

logging.basicConfig(
    filename='app_activity.log',
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s',
    filemode='a'
)

def perform_calculation(a, b):
    logging.info(f"Attempting to divide {a} by {b}...")

    try:
        result = a / b
        logging.info(f"Calculation successful. Result: {result}")
        return result
    except ZeroDivisionError:
        logging.error("Failed to divide: Denominator was zero.")
        return None
    except TypeError:
        logging.error(f"Failed to divide: Invalid types provided (a={type(a)}, b={type(b)}).")
        return None
    logging.shutdown()

print("Check 'app_activity.log' to see the output.")


perform_calculation(10, 2)


perform_calculation(10, 0)


perform_calculation(10, "five")
```

###19.  Write a program that handles both IndexError and KeyError using a try-except block

```python
def access_data(data_list, data_dict, index, key):
    try:

        print(f"List item at index {index}: {data_list[index]}")


        print(f"Dictionary value for key '{key}': {data_dict[key]}")

    except IndexError:
        print(f"Error: Index {index} is out of range for the list.")

    except KeyError:
        print(f"Error: The key '{key}' was not found in the dictionary.")

    except Exception as e:
        print(f"An unexpected error occurred: {e}")


my_list = ["Apple", "Banana", "Cherry"]
my_dict = {"id": 101, "name": "Gemini"}


print("Test 1:")
access_data(my_list, my_dict, 1, "name")

print("\nTest 2 (Triggers IndexError):")
access_data(my_list, my_dict, 10, "name")

print("\nTest 3 (Triggers KeyError):")
access_data(my_list, my_dict, 0, "age")
```

###21.  Write a Python program that reads a file and prints the number of occurrences of a specific word

```python
import os

def count_word_in_file(file_path, target_word):
    count = 0
    target_word = target_word.lower()

    try:
        with open(file_path, 'r') as file:
            for line in file:

                words = line.lower().split()


                count += words.count(target_word)

        print(f"The word '{target_word}' appears {count} times in the file.")
        return count

    except FileNotFoundError:
        print("Error: The file was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")


count_word_in_file('file.txt', 'This')
```

###22.  How can you check if a file is empty before attempting to read its contents

```python
import os

file_path = 'file.txt'

if os.path.exists(file_path):
    if os.path.getsize(file_path) > 0:
        with open(file_path, 'r') as f:
            print(f.read())
    else:
        print("The file is empty.")
else:
    print("File not found.")
```

###23. Write a Python program that writes to a log file when an error occurs during file handling.

```python
import logging
import os


logging.basicConfig(
    filename='file_errors.log',
    level=logging.ERROR,
    format='%(asctime)s - %(levelname)s - %(message)s'
)

def read_important_file(filename):
    try:
        with open(filename, 'r', encoding='utf-8') as file:
            data = file.read()
            print("File read successfully!")
            return data

    except FileNotFoundError:

        logging.error(f"File Not Found: Attempted to access '{filename}'")
        print("Error: The requested file is missing. Check 'file_errors.log' for details.")

    except Exception as e:

        logging.error(f"Unexpected Error with {filename}: {str(e)}")
        print("An unexpected error occurred.")


read_important_file("missing_data.txt")
```
