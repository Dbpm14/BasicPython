1. What are data structures, and why are they important?

    -Data structures in Python are specialized formats or containers used to store and organize data efficiently.They provide various ways to store, access, and manipulate data according to specific needs.

2. Explain the difference between mutable and immutable data types with examples.

    -Mutable Data types are those data types that can be changed/altered after their creation. Ex.- List, Sets.
    Whereas Immutable data types are those which cannnot be changed/altered after their creation. Ex. Tuples, Strings, Frozensets.

3. What are the main differences between lists and tuples in Python?
    
    -The main difference between lists and tuples is that Lists are mutable whereas tuples are immutable.Hence we can use methods like append, extend and remove in lists but not in tuples.


4. Describe how dictionaries store data.

    -Dictionaries are key-value pairs that allows to store and retrieve data based on a unique key. The keys must be immutable, such as strings or numbers.

5. Why might you use a set instead of a list in Python ?

      -A set inherently stores only unique values. If one adds a duplicate element, it is ignored. This makes a set ideal for situations where one must ensure every item is represented only once.

6. What is a string in Python, and how is it different from a list?

    -A String is an immutable sequence of characters, used to store text data.Strings are created by enclosing characters within single quotes ('...'), double quotes ("..."), or triple quotes ("""...""" or '''...''').
    
    While a list is a mutable sequence of ordered elements, which can hold items of different data types.

7. How do tuples ensure data integrity in Python?

    -Tuples ensure data integrity through their characteristic of immutability. Once a tuple is created, its elements cannot be changed, added, or removed, which guarantees that the data remains constant throughout the program's execution.

8.  What is a hash table, and how does it relate to dictionaries in Python?

      
      -A hash table is a data structure that efficiently maps keys to values using a technique called hashing. It allows for fast data retrieval, insertion, and deletion by using a hash function to calculate an index where the corresponding value should be stored.

      The dictionary in Python is a high-level, user-friendly implementation of the hash table concept, providing an intuitive and efficient way to store and retrieve data

10. Explain why strings are immutable in Python.

    -Strings in Python are immutable i.e. they cannot be changed after they are created. Strings are immutable by design to keep them safe, consistent, and efficient. Immutability makes strings hashable.

12. Describe a scenario where using a tuple would be preferable over a list.

    -A scenario where using a tuple would be preferable over a list is when you need to represent a fixed collection of heterogeneous data that should not change after its creation.

13.  How do sets handle duplicate values in Python?

      -Sets are defined as unordered collections of unique elements therefore, they automatically discard or ignore duplicate values

14. How does the “in” keyword work differently for lists and dictionaries?

    -When used with a list, the in keyword checks for the presence of a specific value among the list's elements. The check is a linear search through the elements of the list.

    When used with a dictionary, the in keyword checks for the presence of a specific key. It does not directly check if a specific value exists in the dictionary's values.

15. Can you modify the elements of a tuple? Explain why or why not?

    -No, one cannot modify the elements of a tuple after it has been created because tuples are immutable data structures. Once a tuple is defined, the memory location and values of its elements are set and cannot be changed.

16. What is a nested dictionary, and give an example of its use case?

    -A nested dictionary is a dictionary within another dictionary, allowing for a hierarchical storage of data. The outer dictionary holds keys that map to other dictionaries as their values. This structure is particularly useful for representing complex, multi-layered data structures.
    
    A practical example use case is managing a database of user information, where each user has multiple related attributes

18.  In what situations are lists preferred over dictionaries?


      -Lists are preferred over dictionaries when order matters, one needs to manage sequential data, perform operations on all items, allow duplicates, or need fast index-based access.

19. Why are dictionaries considered unordered, and how does that affect data retrieval?


    -Dictionaries are considered unordered because they store values as key-value pair and not by indexing, hence considered unordered as order of items may not match the insertion order.

    This affects retrieval by making direct indexed access (like dict[0]) impossible, instead, one must use keys (e.g., dict['name']) to fetch values.

20.  Explain the difference between a list and a dictionary in terms of data retrieval.


     -A list retrieves data based on its item's integer index number where 0 denotes the index of the first item whereas a Dictionary retrieves data based on their unique key and not their integer index
