##Theory Question

1. What is Python, and why is it popular
    -Python is a high-level interpreted programming language with quite easy and understandable syntax, resembling the English language. This language has gained immense popularity due to the fact that a newcomer can easily understand the syntax, there are numerous underlying functions with the help of which many tasks can be performed easily, such as Data Science and Artificial Intelligence tasks, and much more.

2. What is an interpreter in Python

    -An interpreter is a software program that runs the program line by line. An interpreter does not behave like a compiler, which translates the entire program into machine code before executing the program. The Python interpreter translates a single command into an intermediate form of bytecode and then immediately executes that command. This is easier to debug, as the program will stop where the bug occurred.

3. What are pre-defined keywords in Python

    -Keywords in Python are reserved words that carry a fixed meaning and functionality. They comprise the basic syntax and structure of the programming language. Examples of keywords in Python include the keyword 'if' used in logical operations, the 'for' keyword used in loops, and the keyword 'import' used in the addition of libraries. They can neither be used as variables nor for any other functionality in a program.

4. Can keywords be used as variable names?

    -Keywords can neither be used for variable names, function names, or any other designation. Doing so will lead to a SyntaxError since the interpreter is looking for an operation related to the specific functions that the keywords can undertake.

6. Why are lists mutable, but tuples immutable?

    -Lists are mutable since they are designed to retain sets of data that may possibly change in size while running a program. Tuples are immutable due to the intended function of holding unchanging sets of data. Using this approach ensures integrity in the data held in a tuple and is so safe for use as constants or dictionary keys to allow for memory optimization in terms of performance.

7. How do “==” and “is” operators differ?

    `==` is a comparison for equality; it will check if the values of two objects are equal. `is` is a comparison for identity; it will check if two variables refer to the same location in memory (in other words, they are the same object).

8. What are logical operators in Python?
    These are used in combination with conditional statements. The common ones are and (it yields True if and only if the two statements are True), or (it yields True if at least one of the statements are True), and not (reverses the boolean output – True becomes False and vice versa).
  
  9. What is type casting in Python?
    -Type Casting
    Type casting is the process whereby a variable can be changed from one type to another. Type casting makes it possible for operations to be carried out correctly even when data is in a different form from what is expected, for example, when a numeric value is converted to an integer to enable calculations.
  
  10. What is the difference between implicit and explicit type casting?
    -“Implicit type casting” is an automatically accomplished operation on Python’s part in order to avoid data loss (for instance, adding an integer to a float will result in an automatically created float). “Explicit Type Casting,” or “type conversion,” can be carried out manually by the programmer using the int(), float(), str() functions in Python.

The conditional statement allows the execution of certain instructions in a program based on a particular qualifying condition that must be satisfied, thus enabling the computer to "make decisions" or manage different situations based on their processes.
  
  12. What does the elif statement do?
    -The elif part (short form of "else if") enables you to test multiple conditions for Truth. It is sandwiched between the if statement and else. When the first if condition is evaluated as False, it then proceeds to evaluate the condition in the elif, then moves on if it is also False. There is no need for nesting.

```python
# 1. Working with a List (Mutable)

my_list = [10, 20, 30]
my_list[0] = 99
print("Updated List:", my_list)

# 2. Working with a Tuple (Immutable)

my_tuple = (10, 20, 30)
my_tuple[0] = 99
print(my_tuple)
```

```python
# Mutable behavior (List)
original_list = [1, 2, 3]
copied_list = original_list
copied_list.append(4)

print("Original List:", original_list)
print("Copied List:", copied_list)

# Immutable behavior (String)
original_string = "Hello"
new_string = original_string
new_string = new_string + " World"

print("Original String:", original_string)
print("New String:", new_string)
```

```python
total_sum = 0
for i in range(1, 51):
    if i % 2 == 0:
        total_sum += i

print("Sum of even numbers 1-50:", total_sum)
```

```python
text = input("Enter a string: ")
reversed_text = ""
index = len(text) - 1

while index >= 0:
    reversed_text += text[index]
    index -= 1

print("Reversed string:", reversed_text)
```
